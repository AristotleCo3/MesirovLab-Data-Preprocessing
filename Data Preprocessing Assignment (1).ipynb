1. Write a Python function with the following functional spec: 
- Inputs:
  - a GenePattern GCT file
  - An optional summary parameter, boolean, with a default value of False.
- Output: a Pandas DataFrame containing the contents of the input file
   
The function should read the GCT file into a Pandas DataFrame. You do not need to worry about validating the file format and should make use of the file reading functions available through Pandas.
If the optional summary parameter is passed and equals True, the function should print the number of rows and columns imported.
Discard the first 2 lines of the GCT file, which are headers that do not contain biological information.
Make sure that the first 2 columns, Name and Description, are indexes of the DataFrame and not part of the data.














```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def read_file(GCT_file, summary=False): #read GCT file into a pandas dataframe
    df = pd.read_csv(GCT_file, skiprows=2, sep='\t') #delete first two rows which don't contain biological info
    #I researched a bit and found that GCT data's column are separated by tab delimiter, so I used sep='\t'to let pandas recognize different columns and avoiding potential errors.
    if summary == True:
        print('The number of rows imported is ' + str(df.shape[0]))
        print('The number of columns imported is ' + str(df.shape[1]))
    
    df = df.set_index(['Name', 'Description']) #set Name and Description as indexes
    
    return df
```

4. Using the data in the DataFrame from #2b, print out a DataFrame containing one row for each sample, with the following columns computed:
- Sample name
- Mean
- Median
- Standard deviation

```python
def statistics(df): #create a function that put sample name, mean, median, and standard deviation of each column into a dataframe
    
    sample_name = df.columns
    mean = df.mean(axis=0)
    median = df.median(axis=0)
    std = df.std(axis=0)
    
    stats_df = pd.DataFrame({
        'Sample name':sample_name,
        'Mean':mean,
        'Median':median,
        'Standard deviation':std
    })
    
    return stats_df

statistics(GCT_df)
```
